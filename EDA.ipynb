## 인접 지역 종합

```python
total_travel_purpose = pd.merge(yeongju_travel_purpose, danyang_travel_purpose, on = '방문목적', how = 'outer', suffixes = ('_영주', '_단양'))
total_travel_purpose = pd.merge(total_travel_purpose, andong_travel_purpose, on = '방문목적', how = 'outer')
total_travel_purpose = pd.merge(total_travel_purpose, yecheon_travel_purpose, on = '방문목적', how = 'outer', suffixes = ('_안동', '_예천'))
total_travel_purpose = pd.merge(total_travel_purpose, bonghwa_travel_purpose, on = '방문목적', how = 'outer')
total_travel_purpose = pd.merge(total_travel_purpose, taebaek_travel_purpose, on = '방문목적', how = 'outer', suffixes = ('_봉화', '_태백'))
total_travel_purpose.fillna(0, inplace = True)
total_travel_purpose[['응답횟수_영주', '응답횟수_단양', '응답횟수_안동', '응답횟수_예천', '응답횟수_봉화', '응답횟수_태백']] = total_travel_purpose[['응답횟수_영주', '응답횟수_단양', '응답횟수_안동', '응답횟수_예천', '응답횟수_봉화', '응답횟수_태백']].astype(int)
total_travel_purpose.set_index('방문목적', inplace = True)
total_travel_purpose.columns = ['영주', '단양', '안동', '예천', '봉화', '태백']
total_travel_purpose.loc['total'] = total_travel_purpose.sum()
total_travel_purpose
```

```python
# 비율로 변환
for local in ['영주', '단양', '안동', '예천', '봉화', '태백']:
    total_travel_purpose[local] = total_travel_purpose[local] / total_travel_purpose.loc['total', local] * 100
    
total_travel_purpose.drop('total', inplace = True)
total_travel_purpose
```

```python
total_travel_purpose
```

```python
# 전체 역사 유적지 방문
plt.figure(figsize=(12, 7))

sns.barplot(
    x = total_travel_purpose.columns,
    y = total_travel_purpose.loc['쇼핑'],
    hue = total_travel_purpose.columns,
    palette = 'Accent',
    legend = False
)

plt.yticks(fontsize = 15, fontweight = 'bold')
plt.title(' ', pad = 50)
plt.xlabel(' ', labelpad = 50)
plt.ylabel(' ', labelpad = 50)

plt.gca().yaxis.set_major_formatter(PercentFormatter(100, decimals = 0))

plt.gca().spines['right'].set_visible(False) #오른쪽 테두리 제거
plt.gca().spines['top'].set_visible(False) #위 테두리 제거
plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 제거
plt.gca().set_xticklabels([])  # X축 레이블을 빈 리스트로 설정하여 제거
plt.gca().axes.xaxis.set_visible(False) #x축 제거

plt.show()
```

# 각 지역별 방문자 나이대

```python
yeongju_traveler_age = get_traveler_age('영주')
danyang_traveler_age = get_traveler_age('단양')
andong_traveler_age = get_traveler_age('안동')
yecheon_traveler_age = get_traveler_age('예천')
bonghwa_traveler_age = get_traveler_age('봉화')
taebeak_traveler_age = get_traveler_age('태백')

total_traveler_age = pd.merge(yeongju_traveler_age, danyang_traveler_age, on = '연령대', how = 'outer')
total_traveler_age = pd.merge(total_traveler_age, andong_traveler_age, on = '연령대', how = 'outer')
total_traveler_age = pd.merge(total_traveler_age, yecheon_traveler_age, on = '연령대', how = 'outer')
total_traveler_age = pd.merge(total_traveler_age, bonghwa_traveler_age, on = '연령대', how = 'outer')
total_traveler_age = pd.merge(total_traveler_age, taebeak_traveler_age, on = '연령대', how = 'outer')
total_traveler_age.fillna(0, inplace = True)
total_traveler_age.set_index('연령대', inplace = True)
total_traveler_age[['영주', '단양', '안동', '예천', '봉화', '태백']] = total_traveler_age[['영주', '단양', '안동', '예천', '봉화', '태백']].astype(int)
total_traveler_age
```

```python
# 비율로 변환
total_traveler_age.loc['total'] = total_traveler_age.sum()

for local in ['영주', '단양', '안동', '예천', '봉화', '태백']:
    total_traveler_age[local] = total_traveler_age[local] / total_traveler_age.loc['total', local] * 100
    
total_traveler_age.drop('total', inplace = True)
total_traveler_age
```

```python
# 데이터 프레임을 긴 형식으로 변환
melted_data = total_traveler_age.reset_index().melt(id_vars='연령대')
melted_data.columns = ['AgeGroup', 'Category', 'Value']

# seaborn으로 막대 그래프 그리기
plt.figure(figsize=(10, 6))
sns.barplot(
    x='AgeGroup', 
    y='Value', 
    hue='Category',  # 여러 카테고리로 구분하여 표현
    data=melted_data,
    palette='Accent'
)

plt.legend(title='Category')
plt.show()
```

```python
plt.figure(figsize = (20, 15))

for idx, purpose in enumerate(total_traveler_age.index):
    plt.subplot(3, 2, idx + 1)
    sns.barplot(
        x = total_traveler_age.columns,
        y = total_traveler_age.loc[purpose],
        hue = total_traveler_age.columns,
        palette = 'Accent'
    )
    plt.xlabel(None)
    plt.ylabel(None)
    plt.title(purpose)
```
